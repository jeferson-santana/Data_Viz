```python
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

Link para Biblioteca: https://seaborn.pydata.org/index.html

Pegando dados do Seaborn :https://github.com/mwaskom/seaborn-data

```python
#dados link: https://github.com/mwaskom/seaborn-data
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.info()
```

```python
#bar plot, que dia gasta mais
sns.barplot(data=tips,y='total_bill',x='day')
```

```python
#que dia tem mais gorgeta
sns.barplot(data=tips,y='tip',x='day')
```

```python
#Plot por linhas
sns.stripplot(x="day", y="total_bill", data=tips);
```

```python
sns.stripplot(x="day", y="total_bill", data=tips, jitter=True);
```

```python
#bar plot diferenciando dados
sns.barplot(data=tips,y='tip',x='day',hue='sex')
```

```python
#countPlot
sns.countplot(data=tips,x='sex')
```

```python
#box plot (diagrama de caixa)
sns.boxplot(data=tips,x='sex',y='total_bill')
```

```python
import matplotlib.pyplot as plt
```

```python
#alterando tamanho da imagem
plt.figure(figsize=(5,10))
sns.boxplot(data=tips,x='sex',y='total_bill')
```

```python
#diferenciando dados
plt.figure(figsize=(5,7))
sns.boxplot(x="day", y="total_bill", hue="time", data=tips);
```

```python
#violino plot
sns.violinplot(data=tips,x="day", y="total_bill",hue='sex',split=True);
```

```python
#verificando a distribuição
sns.distplot(tips['total_bill']);
```

```python
#verificando a distribuição aumentando o bins
sns.distplot(tips['total_bill'],bins=100);
```

```python
#verificando a distribuição
sns.distplot(tips['tip'],kde=False);
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
#Retirando o kde e colocando o rug
sns.distplot(tips['tip'], kde=False, rug=True);
```

```python
#Verificando corelação
sns.lmplot(data=tips,x='total_bill',y='tip')
```

```python
#alterando a ordem
sns.lmplot(data=tips,x='total_bill',y='tip',order=3)
```

```python
#sem a regressão
sns.lmplot(data=tips,x='total_bill',y='tip',fit_reg=False)
```

```python
#com o histograma
sns.jointplot(x="total_bill", y="tip", data=tips);
```

```python
#Com o histograma bonito
sns.jointplot(x="total_bill", y="tip", data=tips,kind="hex");
```

```python
#verificando todos os dados
sns.pairplot(tips);
```

```python
iris = sns.load_dataset('iris')
```

```python
sns.pairplot(iris);
```

```python
#com o os clusters
g = sns.PairGrid(iris)
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.kdeplot, cmap="Blues_d", n_levels=6);
```

```python
#varios plots
g = sns.PairGrid(tips,
                 x_vars=["smoker", "time", "sex"],
                 y_vars=["total_bill", "tip"],
                 aspect=.75, size=3.5)
g.map(sns.barplot, palette="pastel");
```
